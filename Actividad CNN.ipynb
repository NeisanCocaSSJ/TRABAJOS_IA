````python
from keras.datasets import fashion_mnist
(train_X,train_Y), (test_X,test_Y) = fashion_mnist.load_data()

# Commented out IPython magic to ensure Python compatibility.
import numpy as np
from keras.utils import to_categorical
import matplotlib.pyplot as plt

#%matplotlib inline
print('Training data shape : ', train_X.shape, train_Y.shape)
print('Testing data shape:', test_X.shape, test_Y.shape)

#Find the unique numbers from the train labels
classes = np.unique(train_Y)
nClasses = len(classes)
print('Total number of outputs: ', nClasses)
print('Output classes : ', classes)
plt.figure(figsize=[5,5])

#Display the first image in training data
plt.subplot(121)
plt.imshow(train_X[0,:,:], cmap='gray')
plt.title("Ground Truth : {}".format(train_Y[0]))

# Display the first image in testing data
plt.subplot(122)
plt.imshow(test_X[0,:,:], cmap='gray')
train_X = train_X.reshape(-1, 28,28, 1)
test_X = test_X.reshape(-1, 28,28, 1)
train_X.shape, test_X.shape
train_X = train_X.astype('float32')
test_X = test_X.astype('float32')
train_X = train_X/255.
test_X = test_X/255.

# Change the labels from categorical to one-hot encoding
train_Y_one_hot = to_categorical(train_Y)
test_Y_one_hot = to_categorical(test_Y)

# Display the change for category label using one-hot encoding
print('Original label:', train_Y[0])
print('After conversion to one-hot:', train_Y_one_hot[0])

from sklearn.model_selection import train_test_split
train_X,valid_X,train_label,valid_label = train_test_split(train_X, train_Y_one_hot, test_size=0.2, random_state=13)
train_X.shape,valid_X.shape, train_label.shape,valid_label.shape

import keras
from keras.models import Sequential, Model
from tensorflow.keras import Input
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from keras.layers import LeakyReLU

batch_size = 64
epochs = 5 #20
num_classes = 10
fashion_model = Sequential()
fashion_model.add(Conv2D(32, kernel_size=(3,3), activation='linear',input_shape=(28,28,1), padding='same'))
fashion_model.add(LeakyReLU(alpha=0.1))
fashion_model.add(MaxPooling2D((2, 2), padding='same'))
fashion_model.add(Conv2D(64, (3, 3), activation='linear', padding='same'))
fashion_model.add(LeakyReLU(alpha=0.1))
fashion_model.add(MaxPooling2D(pool_size=(2, 2), padding='same'))
fashion_model.add(Conv2D(128, (3, 3), activation='linear',padding='same'))
fashion_model.add(LeakyReLU(alpha=0.1))
fashion_model.add(MaxPooling2D(pool_size=(2, 2), padding='same'))
fashion_model.add(Flatten())
fashion_model.add (Dense(128, activation='linear'))
fashion_model.add(LeakyReLU(alpha=0.1))
fashion_model.add(Dense(num_classes, activation='softmax'))
fashion_model.compile(loss=keras.losses.categorical_crossentropy, optimizer=keras.optimizers.Adam(), metrics=['accuracy'])
fashion_model.summary()
fashion_train = fashion_model.fit(train_X, train_label, batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(valid_X, valid_label))
test_eval = fashion_model.evaluate(test_X, test_Y_one_hot, verbose=0)

print('Test loss:', test_eval[0])
print('Test accuracy:', test_eval[1])
accuracy = fashion_train.history['accuracy']
val_accuracy = fashion_train.history['val_accuracy']
loss = fashion_train.history['loss']
val_loss = fashion_train.history['val_loss']
epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title ('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
predicted_classes = fashion_model.predict(test_X)
predicted_classes = np.argmax(np.round(predicted_classes), axis=1)
predicted_classes.shape, test_Y.shape

correct = np.where(predicted_classes==test_Y)[0]
print ('Found %d correct labels' % len(correct))
for i, correct in enumerate(correct[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(test_X[correct].reshape(28,28), cmap='gray', interpolation='none')
    plt.title("Predicted {}, Class {}".format(predicted_classes[correct], test_Y[correct]))
    plt.tight_layout()

incorrect = np.where(predicted_classes!=test_Y)[0]
print ('Found %d incorrect labels' % len(incorrect))
for i, incorrect in enumerate(incorrect[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(test_X[incorrect].reshape(28,28), cmap='gray', interpolation='none')
    plt.title("Predicted {}, Class {}".format(predicted_classes[incorrect], test_Y[incorrect]))
    plt.tight_layout()
````

## Explicación técnica de líneas clave en la capa densa de una CNN.
A continuación, se explica el propósito de dos instrucciones frecuentes utilizadas en la
parte densa de una Red Neuronal Convolucional:

#### 1. model.add(Dropout(0.3))
La capa Dropout es una técnica de regularización que ayuda a prevenir el sobreajuste.
Durante el entrenamiento, esta capa desactiva aleatoriamente el 30% de las neuronas
en la capa donde se aplica, evitando así que el modelo se vuelva demasiado
dependiente de ciertos nodos. Esto mejora la capacidad de generalización del modelo
sobre datos no vistos.

#### 2. model.add(LeakyReLU(alpha=0.1))
LeakyReLU es una función de activación que, a diferencia de ReLU tradicional, permite
pasar un pequeño valor cuando la entrada es negativa. En este caso, 'alpha=0.1' indica
que, para entradas negativas, se permite pasar el 10% del valor en lugar de cero. Esto
evita el problema de 'neuronas muertas' y mejora el aprendizaje al mantener el flujo del
gradiente durante el entrenamiento, especialmente útil en redes profundas.
Ambas líneas son utilizadas comúnmente en la parte final de una CNN, justo después
de la capa Flatten o antes de la capa de salida, para fortalecer la estabilidad y
capacidad predictiva del modelo.

4. Explica con tus palabras el propósito de cada una de las capas en una CNN.
Capa de Conv2D(Convulcion):

- Capa Conv2D (Convolución): Es el corazón de la CNN. Aplica filtros (kernels) a la imagen de entrada para detectar patrones locales, como bordes, texturas y formas. La capa "escanea" la imagen y produce mapas de características que representan dónde se encuentran estos patrones.

- Capa de Activación (LeakyReLU/ReLU): Se aplica después de la convolución para introducir no linealidad en el modelo. Esto permite que la red aprenda relaciones complejas. LeakyReLU, en particular, permite un flujo de gradiente pequeño para entradas negativas, evitando el problema de las "neuronas muertas" que puede ocurrir con ReLU tradicional.

- Capa MaxPooling2D (Pooling/Agrupación): Reduce las dimensiones espaciales (alto y ancho) de los mapas de características. Esto reduce el número de parámetros y el costo computacional, y ayuda a hacer que las características detectadas sean más robustas a pequeños desplazamientos en la imagen.

- Capa Flatten: Transforma la salida 3D (alto x ancho x profundidad) de las capas convolucionales/pooling en un vector 1D. Esto es necesario para conectar la parte de extracción de características (Conv/Pool) con la parte de clasificación (capas densas).

- Capa Dense (Densa/Fully Connected): Realiza la clasificación final. Recibe el vector de características de Flatten y aprende las relaciones de alto nivel para mapear las características extraídas a la probabilidad de cada clase (por ejemplo, "camiseta", "pantalón", etc.).

- Capa Dropout: Una capa de regularización que desactiva aleatoriamente algunas neuronas durante el entrenamiento. Su propósito es prevenir el sobreajuste, forzando a la red a no depender de ninguna neurona o característica específica.

- Capa Softmax (como activación de la capa de salida): Se utiliza en la capa de salida para problemas de clasificación multiclase. Convierte los valores numéricos de la capa densa en una distribución de probabilidad, donde la suma de las probabilidades para todas las clases es igual a 1. El valor más alto indica la clase predicha.

````python
from keras.datasets import fashion_mnist # Carga el dataset (Conjunto de Datos)
(train_X,train_Y), (test_X,test_Y) = fashion_mnist.load_data() # División inicial de Datos (Entrenamiento y Prueba)

# Commented out IPython magic to ensure Python compatibility.
import numpy as np
from keras.utils import to_categorical # Utilidad para One-Hot Encoding
import matplotlib.pyplot as plt

#------------------------------------------------------------------------------------
from sklearn.model_selection import train_test_split # Herramienta para crear el Conjunto de Validación
from sklearn.metrics import classification_report # Métrica de Evaluación Detallada (Reporte por Clase)
#------------------------------------------------------------------------------------

#%matplotlib inline
# Se imprime la forma de los tensores de datos (Inspección de Datos)
print('Training data shape : ', train_X.shape, train_Y.shape) 
print('Testing data shape:', test_X.shape, test_Y.shape)

# Find the unique numbers from the train labels
classes = np.unique(train_Y)
nClasses = len(classes)
print('Total number of outputs: ', nClasses) # Número de Clases a clasificar
print('Output classes : ', classes)
plt.figure(figsize=[5,5])

# Bloque de Visualización de Datos (Muestra la imagen antes del pre-procesamiento)
plt.subplot(121) 
plt.imshow(train_X[0,:,:], cmap='gray')
plt.title("Ground Truth : {}".format(train_Y[0]))

plt.subplot(122)
plt.imshow(test_X[0,:,:], cmap='gray')
plt.title("Ground Truth : {}".format(test_Y[0])) 
plt.show() 

# Transformación de las Imágenes (Reestructuración y Normalización)
train_X = train_X.reshape(-1, 28,28, 1) # Reshape para CNN (Añade el canal de color '1')
test_X = test_X.reshape(-1, 28,28, 1)
train_X.shape, test_X.shape
train_X = train_X.astype('float32') # Tipado para cálculos de punto flotante
test_X = test_X.astype('float32')
train_X = train_X/255. # Normalización (Escalado de píxeles a [0, 1])
test_X = test_X/255.

# Change the labels from categorical to one-hot encoding
train_Y_one_hot = to_categorical(train_Y) # One-Hot Encoding de etiquetas
test_Y_one_hot = to_categorical(test_Y)

print('Original label:', train_Y[0])
print('After conversion to one-hot:', train_Y_one_hot[0])

# División para el Conjunto de Validación (separa una parte para monitoreo durante el fit)
train_X,valid_X,train_label,valid_label = train_test_split(train_X, train_Y_one_hot, test_size=0.2, random_state=13)
train_X.shape,valid_X.shape, train_label.shape,valid_label.shape

import keras
from keras.models import Sequential # Define un Modelo Secuencial (stack de capas)
from keras.layers import Dense, Dropout, Flatten # Importa Capa Densa, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D # Importa Capa Convolucional y Max Pooling
from keras.layers import LeakyReLU # Importa la Función de Activación LeakyReLU
from tensorflow.keras.layers import BatchNormalization
#---------------------------------------------------------------------------------
from tensorflow.keras.optimizers import RMSprop # Optimizador (Estrategia de Actualización de Pesos)
#---------------------------------------------------------------------------------
from keras.layers import LeakyReLU

# -----------------------------------------------------
# Definición de Hiperparámetros
# -----------------------------------------------------
batch_size = 128 # Batch Size (Tamaño del Lote: estabilidad del gradiente)
epochs = 10 # Épocas (Número de Pasadas completas al dataset)
num_classes = 10

# ---------------------------------------------------
# Definición de Arquitectura (Extracción de Características)
# ---------------------------------------------------

fashion_model = Sequential()

# Primera Capa: Convolución (Extracción de Características)
fashion_model.add(Conv2D(32, kernel_size=(3,3), activation='linear',input_shape=(28,28,1), padding='same')) # Conv2D y 32 Filtros
fashion_model.add(LeakyReLU(alpha=0.1)) # Activación (No-Linealidad)
fashion_model.add(MaxPooling2D((2, 2), padding='same')) # Max Pooling (Reducción de Dimensionalidad)

# Segunda Capa
fashion_model.add(Conv2D(64, (3, 3), activation='linear', padding='same'))
fashion_model.add(LeakyReLU(alpha=0.1))
fashion_model.add(MaxPooling2D(pool_size=(2, 2), padding='same'))

# Tercera Capa
fashion_model.add(Conv2D(128, (3, 3), activation='linear',padding='same'))
fashion_model.add(LeakyReLU(alpha=0.1))
fashion_model.add(MaxPooling2D(pool_size=(2, 2), padding='same'))

#-------------------------------------------------------------------------------------------------------
# Cuarta Capa (Modificación 1.1: Aumenta la Profundidad del Modelo)
fashion_model.add(Conv2D(256, (3, 3), activation='linear',padding='same')) # Aumenta la Capacidad de Abstracción
fashion_model.add(LeakyReLU(alpha=0.1))
fashion_model.add(MaxPooling2D(pool_size=(2, 2), padding='same'))

fashion_model.add(Flatten()) # Flatten (Convierte 3D a 1D para el Clasificador)
# -------------------------------------------------------------------------------------------------------

#--------------------------------------------------------------------------------------------------------
# Capas Densas (Clasificador)
# Capa densa intermedia (Modificaciones 1.2 y 1.3: Aumenta Capacidad de Modelado)
fashion_model.add (Dense(256, activation='relu')) # 256 Unidades y Activación 'relu' (No-Linealidad)
fashion_model.add(LeakyReLU(alpha=0.1))

# Capa Dropout (Modificación 1.4: Regularización)
fashion_model.add(Dropout(0.3)) # Dropout(rate=0.3) (Previene Overfitting / Sobreajuste)
#--------------------------------------------------------------------------------------------------------

# Capa de salida
fashion_model.add(Dense(num_classes, activation='softmax')) # Capa de Salida con Softmax (Probabilidades de Clase)

# Compilación del Modelo
fashion_model.compile(
    loss=keras.losses.categorical_crossentropy, # Función de Pérdida (Mide el Error)
    optimizer=RMSprop(), # Optimizador (Ajusta los Pesos)
    metrics=['accuracy'] # Métrica a Monitorear
)
fashion_model.summary() # Muestra la Arquitectura y el Conteo de Parámetros
fashion_train = fashion_model.fit(
    train_X, train_label, 
    batch_size=batch_size,
    epochs=epochs,
    verbose=1,
    validation_data=(valid_X, valid_label) # Datos de Validación (Monitorea el Desempeño en datos no vistos)
)
test_eval = fashion_model.evaluate(test_X, test_Y_one_hot, verbose=0) # Evaluación Final en el Set de Prueba

print('Test loss:', test_eval[0])
print('Test accuracy:', test_eval[1]) # Precisión Final del Modelo

# ----------------------------------------------------
# Visualización y Reporte de Resultados
# ----------------------------------------------------

# Bloque para generar gráficas de Accuracy y Loss
accuracy = fashion_train.history['accuracy']
val_accuracy = fashion_train.history['val_accuracy'] # Accuracy en el conjunto de Validación
loss = fashion_train.history['loss']
val_loss = fashion_train.history['val_loss'] # Loss en el conjunto de Validación
# El análisis de estas gráficas es clave para evaluar el Overfitting/Underfitting

plt.plot(...)
plt.show()

predicted_classes = fashion_model.predict(test_X)
predicted_classes = np.argmax(np.round(predicted_classes), axis=1) # Decodificación de Predicciones

target_names = ["Class {}".format(i) for i in range(num_classes)]
# Reporte de Clasificación (Métricas detalladas por clase: Precisión, Recall, F1-score)
print(classification_report(test_Y, predicted_classes, target_names=target_names)) 

# Visualización de Aciertos y Errores
# Muestra ejemplos de predicciones correctas e incorrectas
correct = np.where(predicted_classes==test_Y)[0] 
incorrect = np.where(predicted_classes!=test_Y)[0]
````
